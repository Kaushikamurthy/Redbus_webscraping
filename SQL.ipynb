```python
import pandas as pd

# Function to convert 'hh h mm m' format to 'hh:mm:ss' format
def convert_to_hms(duration):
    try:
        hours = 0
        minutes = 0
        seconds = 0
        
        # Check for hours and minutes in the string, extract the numeric values
        if 'h' in duration:
            hours = int(duration.split('h')[0].strip())  # Extract hours
        if 'm' in duration:
            minutes = int(duration.split('m')[0].split()[-1].strip())  # Extract minutes
        
        # Format to hh:mm:ss
        time_str = f'{hours:02}:{minutes:02}:{seconds:02}'
        return time_str
    except Exception as e:
        return "Invalid Format"

# Apply the function to the 'Total_duration' column in-place
db_buses_goa['Total_duration'] = db_buses_goa['Total_duration'].apply(convert_to_hms)

# Display the updated DataFrame to confirm the changes
print(db_buses_goa[['Total_duration']])
```

```python
db_buses_goa["Price"]=db_buses_goa["Price"].astype(str)
```

```python
#convert prices to numeric
db_buses_goa["Price"]=db_buses_goa["Price"].str.replace("INR","")
db_buses_goa["Price"]=db_buses_goa["Price"].astype(float)
db_buses_goa["Price"].fillna(0) 
```

```python
# Convert price column to numeric, setting errors='coerce' to replace invalid values with NaN
db_buses_goa['Price'] = pd.to_numeric(db_buses_goa['Price'], errors='coerce')

# Replace NaN (missing values) in the price column with a default value or None (for NULL in MySQL)
db_buses_goa['Price'] = db_buses_goa['Price'].fillna(0)  # Replace NaN with 0, or use None for NULL
```

```python
db_buses_goa['Price'].unique()
```

```python
# Print rows where 'Price' is NaN
print(db_buses_goa[db_buses_goa['Price'].isnull()])

# Drop rows with NaN in 'Price'
db_buses_goa.dropna(subset=['Price'], inplace=True)
```

```python
import numpy as np
import re

# Function to clean the Ratings column
def clean_rating(value):
    # If the value is one of the invalid ones, return NaN
    if str(value).lower() in ['nan', 'new', ' ', 'none']:
        return np.nan
    # Attempt to extract a numeric value from the string
    match = re.match(r'(\d+\.\d+)', str(value))
    if match:
        return float(match.group(1))
    # If no match is found, return NaN
    return np.nan

# Clean the Ratings column
db_buses_goa['Ratings'] = db_buses_goa['Ratings'].apply(clean_rating)

# Show the cleaned data
print(db_buses_goa[['Ratings']].head())
```

```python
db_buses_goa['Ratings'].unique()
```

```python
db_buses_goa['Seats_Available'].unique()
```

```python
# Fill any missing values (NaN) in the 'Seats_Available' column with the string "not available"
db_buses_goa['Seats_Available']=db_buses_goa['Seats_Available'].fillna("not available")
```

```python
db_buses_goa['Seats_Available'].unique()
```

```python
missing_values =db_buses_goa.isnull().sum()
print("Missing values per column:\n", missing_values)
```

```python
# Remove rows with missing values in any column
db_buses_goa = db_buses_goa.dropna()
```

```python
print("Data after removing rows with missing values:\n", db_buses_goa)
```
