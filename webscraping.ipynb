```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException


# Function to extract route links and route names from the website with pagination handling
from selenium.common.exceptions import NoSuchElementException

from selenium.common.exceptions import ElementClickInterceptedException

# Initialize lists to store the extracted links and route names
def kadamba_route_link(path):
    kadamba_link = []
    kadamba_route = []
 # Loop through the first 5 pages
    for i in range(1, 6):
        paths = driver_goa.find_elements(By.XPATH, path) # Find all elements matching the provided XPath
        
 # Extract the link and route name from each element
        for link in paths:
            kadamba_link.append(link.get_attribute("href"))
            kadamba_route.append(link.text)

        try:
            # Wait for the pagination control
            page_control = WebDriverWait(driver_goa, 10).until(
                EC.presence_of_element_located((By.XPATH, '//div[@class="DC_117_paginationTable"]'))
            )
            next_button = page_control.find_element(
                By.XPATH, f'//div[@class="DC_117_pageTabs " and text()={i + 1}]'
            )

            # Scroll to the next button to ensure visibility
            driver_goa.execute_script("arguments[0].scrollIntoView();", next_button)
            time.sleep(1)  # Wait for any animations to finish

            try:
                next_button.click()
            except ElementClickInterceptedException:
                print("Next button is obscured. Attempting workaround...")
                
                driver_goa.execute_script("arguments[0].click();", next_button)

        except NoSuchElementException:
            print(f"No more pages to paginate at step {i}")
            break

    return kadamba_link, kadamba_route

# Run the function
kadamba_link, kadamba_route = kadamba_route_link("//a[@class='route']")

driver_goa.quit()
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementClickInterceptedException

# Function to extract route links and route names from the website 
def up_route_link(path):
    up_link = []
    up_route = []
    
 #Loop through the first 6 pages
    for i in range(1, 6):
        paths = driver_up.find_elements(By.XPATH, path)

        for link in paths:
            up_link.append(link.get_attribute("href"))
            up_route.append(link.text)

        try:
            # Wait for the pagination control
            page_control = WebDriverWait(driver_up, 10).until(
                EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]'))
            )
            next_button = page_control.find_element(
                By.XPATH,  f'//div[@class="DC_117_pageTabs " and text()={i+1}]'
            )

            # Scroll to the next button to ensure visibility
            driver_up.execute_script("arguments[0].scrollIntoView();", next_button)
            time.sleep(1)  # Wait for any animations to finish

            try:
                next_button.click()
            except ElementClickInterceptedException:
                print("Next button is obscured. Attempting workaround...")
                
                driver_up.execute_script("arguments[0].click();", next_button)

        except NoSuchElementException:
            print(f"No more pages to paginate at step {i}")
            break

    return up_link, up_route

# Run the function
up_link, up_route = up_route_link("//a[@class='route']")

driver_up.quit()
```
